# Perceptron example: iris dataset classification

Network: perceptron \
Decision rule: step function \
Error update rule: learning rate * (target - prediction) \
Hyper-plane: step function(transpose(W) * target) = 0

Step 1. Load dataset

```python
import pandas as pd
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header = None)
df.tail()
```

Step 2. Data pre-processing

```python
import matplotlib.pyplot as plt
import numpy as np

Y = df.iloc[0:100, 4]
Y = np.where(Y == 'Iris-setosa',-1,1)
X = df.iloc[0:100,[0,2]].values

plt.scatter(X[:50,0],X[:50,1],color='red',marker='o',label='setosa')
plt.scatter(X[50:100,0],X[50:100,1],color='blue',marker='x',label='versicolor')
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc = 'upper left')
plt.show()
```

Step 3. Training

```python
import func

ppn0 = func.Perceptron(eta = 10, n_iter = 10)
ppn0.fit(X,Y)
ppn1 = func.Perceptron(eta = 0.1, n_iter = 10)
ppn1.fit(X,Y)
ppn2 = func.Perceptron(eta = 0.001, n_iter = 10)
ppn2.fit(X,Y)
ppn3 = func.Perceptron(eta = 0.0001, n_iter = 10)
ppn3.fit(X,Y)


plt.plot(range(1,len(ppn0.errors_) + 1),ppn0.errors_,color='red'  ,label = 'learning rate = 10')
plt.plot(range(1,len(ppn1.errors_) + 1),ppn1.errors_,color='green',label = 'learning rate = 0.1')
plt.plot(range(1,len(ppn2.errors_) + 1),ppn2.errors_,color='blue' ,label = 'learning rate = 0.001')
plt.plot(range(1,len(ppn3.errors_) + 1),ppn3.errors_,color='black',label = 'learning rate = 0.0001')
plt.xlabel('Epochs')
plt.ylabel('Number of errors')
plt.legend(loc = 'upper right')
plt.show()
```

Step 3. Inference (=Prediction)

```python
x1_min, x1_max = X[:,0].min() - 1, X[:,0].max() + 1
x2_min, x2_max = X[:,1].min() - 1, X[:,1].max() + 1
xx1,xx2 = np.meshgrid(np.arange(x1_min,x1_max,0.02),np.arange(x2_min,x2_max,0.02))

Z = ppn1.prediction(np.array([xx1.ravel(),xx2.ravel()]).T)
```

```python
for i in range(10):
    print(str(X[i]) + " -> " + str(Z[i]) + " , Ground truth = " + str(Y[i]))
```

Step 4. Visualization (For Performance analysis)

```python
from Visualization import plot_decision_regions
plot_decision_regions(X,Y,Z)

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc = 'upper left')
plt.show()
```

Appendix. poor classifier visualization

```python
Z_poor = ppn3.prediction(np.array([xx1.ravel(),xx2.ravel()]).T)
plot_decision_regions(X,Y,Z_poor)

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc = 'upper left')
plt.show()
```
